# Intro to Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./data/Advertising.csv")
```

```python
df.head()
```

### Exploring whether there is a relationship between the total advertising costs and the sales

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
sns.scatterplot(x='total_spend', y='sales', data=df)
```

## Least Squares Line

```python
sns.regplot(x='total_spend', y='sales', data=df)
```

```python
X = df['total_spend']
y = df['sales']

# Determine line of best fit (1st Order Polynomial)
poly = np.polyfit(X, y, 1)
poly
```

```python
m = poly[0] # slope
b = poly[1] # y-int

# 
potential_spend = np.linspace(0, 500, 100)
predicted_sales = m * potential_spend + b
```

```python
sns.scatterplot(x='total_spend',y='sales',data=df)
plt.plot(potential_spend,predicted_sales,color='red')
```

### If we spend $256 for the next advertising campaign, how many units do we expect to sell?

```python
spend = 256
expected_sales = m * spend + b
```
